```python
%matplotlib inline 

from pathlib import Path
from ocr.config import PNG_DIR, TMP_DIR
import cv2
```

```python
image_file = Path(PNG_DIR).joinpath('offre.png')
img = cv2.imread(image_file)
```

```python
from ocr.core import display_image_inline
display_image_inline(image_file)
```

### 01: Inverted Image

```python
inverted_img = cv2.bitwise_not(img)
inverted_img_path = Path(TMP_DIR).joinpath('inverted_img.png')
cv2.imwrite(inverted_img_path, inverted_img)
```

```python
display_image_inline(inverted_img_path)
```

### 03: Binarization

```python
def grayscale(image):
    return cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
gray_img_path = Path(TMP_DIR).joinpath('gray_img.png')
gray_image = grayscale(img)
cv2.imwrite(gray_img_path, gray_image)
display_image_inline(gray_img_path)
```
