```python
%matplotlib inline 

from pathlib import Path
from ocr.config import PNG_DIR, TMP_DIR
import cv2
from ocr.core import display_image_inline
```

```python
original_img_path = Path(PNG_DIR).joinpath('carwash.png')
original_img = cv2.imread(original_img_path)
display_image_inline(original_img_path)
```

### 01: Inverted Image

```python
inverted_img = cv2.bitwise_not(original_img)
inverted_img_path = Path(TMP_DIR).joinpath('inverted_img.png')
cv2.imwrite(inverted_img_path, inverted_img)
display_image_inline(inverted_img_path)
```

### 03: Binarization

```python
def grayscale(image):
    return cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
gray_image_path = Path(TMP_DIR).joinpath('gray_img.png')
gray_image = grayscale(original_img)
cv2.imwrite(gray_image_path, gray_image)
display_image_inline(gray_image_path)
```

```python
bw_image_path = Path(TMP_DIR).joinpath('bw_img.png')
thresh, bw_image = cv2.threshold(gray_image, 100, 10000, cv2.THRESH_BINARY)
cv2.imwrite(bw_image_path, bw_image)
display_image_inline(bw_image_path)
```
